```python
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
print(r.text)

```

```python
import json
data = json.loads(r.text)

k = {}
for d in data:
    print(d['title'])
    k=d
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

for d in data:
    print(d['title'])
    print(d['createdAt'])
    print(d['commentCount'])
    print(d['likeCount'])
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

len_p = len(data)
commant_p = 0
like_p = 0
for d in data:
    commant_p += d['commentCount']
    like_p += d['likeCount']

ave_commant_p = commant_p/len_p
ave_like_p = like_p/len_p

print("ave_commant_p = {}".format(ave_commant_p))
print("ave_like_p = {}".format(ave_like_p))

no_p = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
#print(no_p.text)
data_no_p = json.loads(no_p.text)

len_no_p = len(data_no_p)
commant_no_p = 0
like_no_p = 0
for d in data_no_p:
    commant_no_p += d['commentCount']
    like_no_p += d['likeCount']

ave_commant_no_p = commant_no_p/len_no_p
ave_like_no_p = like_no_p/len_no_p

print("ave_commant_no_p = {}".format(ave_commant_no_p))
print("ave_like_no_p = {}".format(ave_like_no_p))
```
